## Results

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('/content/yolov7/runs/train/yolov7-custom3/results.png')
height, width = img.shape[:2]
resized_image = cv2.resize(img,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

fig = plt.gcf()
fig.set_size_inches(18, 10)
plt.axis("off")
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def imShow(imname):
    import cv2
    import matplotlib.pyplot as plt
    %matplotlib inline

    img = cv2.imread(imname)
    height, width = img.shape[:2]
    resized_image = cv2.resize(img,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

    fig = plt.gcf()
    fig.set_size_inches(9,5)
    plt.axis("off")
    plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
import os
import sys

folder = '/content/yolov7/runs/detect/exp'
for filename in os.listdir(folder)[0:10]:
    #img = cv2.imread(os.path.join(folder,filename))
    if filename is not None:
        imShow(os.path.join(folder,filename))
```
